#### 3. 자료구조(List, Dict, Tuple) 처리하기

```python
# 문제 3-1: 리스트에 숫자를 저장하고 합계와 평균 구하기

number = [10, 20, 30,40, 50]

print(f'숫자들: {number}')
print(f'합계: {sum(number)}')
print(f'평균: {sum(number) / len(number)}')
```

```python
# 문제 3-2: 과일 이름 담긴 리스트에서 특정 과일 검색

fruit = ['사과','바나나','오렌지','포도','딸기']
search = input("찾을 과일을 입력하세요: ")

print(f'과일 목록: {fruit}')
print(f'찾을 과일을 입력하세요: {search}')

if search in fruit:
    print(f"'{search}'가 목록에 있습니다!")
else:
    print(f"'{search}'가 목록에 없습니다.")
```

```python
# 문제 3-3: 성적을 딕셔너리에 저장하고 평균 점수 계산

grade = {
    "김철수": 85,
    "이영희": 92,
    "박민수": 78,
    "최수진": 95
}

print(f"학생 성적: ")
for name, score in grade.items():
    print(f'{name}: {score}점')

average = sum(grade.values()) / len(grade)
print(f'평균 점수: {average}점')
```
