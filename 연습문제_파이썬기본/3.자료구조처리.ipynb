#### 3. 자료구조(List, Dict, Tuple) 처리하기

```python
# 문제 3-1: 리스트에 숫자를 저장하고 합계와 평균 구하기

number = [10, 20, 30,40, 50]

print(f'숫자들: {number}')
print(f'합계: {sum(number)}')
print(f'평균: {sum(number) / len(number)}')
```

```python
# 문제 3-2: 과일 이름 담긴 리스트에서 특정 과일 검색

fruit = ['사과','바나나','오렌지','포도','딸기']
search = input("찾을 과일을 입력하세요: ")

print(f'과일 목록: {fruit}')
print(f'찾을 과일을 입력하세요: {search}')

if search in fruit:
    print(f"'{search}'가 목록에 있습니다!")
else:
    print(f"'{search}'가 목록에 없습니다.")
```

```python
# 문제 3-3: 성적을 딕셔너리에 저장하고 평균 점수 계산

grade = {
    "김철수": 85,
    "이영희": 92,
    "박민수": 78,
    "최수진": 95
}

print(f"학생 성적: ")
for name, score in grade.items():
    print(f'{name}: {score}점')

average = sum(grade.values()) / len(grade)
print(f'평균 점수: {average}점')
```

```python
# 문제 3-4: 좌표를 튜플로 저장하고 두 점 사이 거리 계산

n1 = (0, 0)
n2 = (3, 4)
distance = ((n2[0]-n1[0])**2 + (n2[1]-n1[1])**2)**0.5

print(f'점1: {n1}')
print(f'점2: {n2}')
print(f'두 점 사이의 거리: {distance}')
```

```python
# 문제 3-5: 딕셔너리에 상품을 추가하고 총가격 계산
shopping = {
    "사과": [2, 1000],
    "바나나": [3, 800],
    "오렌지": [1, 1500]
}

total_price = 0

print(f'쇼핑 카트: ')
for product, info in shopping.items():
    price = info[0]*info[1]
    print(f'{product}: {info[0]}개 (개당 {info[1]}원) = {price}원')
    
    total_price += price

print(f'총 가격: {total_price}원')
```

```python
# 문제 3-6: 리스트에서 최댓값, 최솟값, 두번째로 큰 값 찾기

num = [15, 3, 27, 8, 19, 12, 31]

print(f'숫자 목록: {num}')
print(f'최댓값: {max(num)}')
print(f'최솟값: {min(num)}')

num.sort(reverse=True)
print(f'두 번째로 큰 값: {num[1]}')
```

```python
# 문제 3-7: 리스트에서 가장 긴 단어와 가장 짧은 단어 찾기

word_list = ['cat', 'elephant', 'dog', 'butterfly', 'ant']
long_word = max(word_list, key=len)
small_word = min(word_list, key=len)

print(f'단어 목록: {word_list}')
print(f'가장 긴 단어: {long_word} ({len(long_word)}글자)')
print(f'가장 짧은 단어: {small_word} ({len(small_word)}글자)')
```

```python
# 문제 3-8: 중복된 요소 제거하고 정렬

origin = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
new = list(set(origin))
new.sort(reverse=False)
print(f'원본 리스트: {origin}')
print(f'중복 제거 후: {new}')
```

```python
# 문제 3-9: 두 개의 리스트 병합하고 공통 요소 찾기

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
new_list = list(set(list1 + list2))
common = []
for x in list1:
    if x in list2:
        common.append(x)

print(f'리스트1: {list1}')
print(f'리스트2: {list2}')
print(f'병합된 리스트: {new_list}')
print(f'공통 요소: {common}')
```
