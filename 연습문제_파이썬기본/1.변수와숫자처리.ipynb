#### 1. 변수와 숫자 처리하기

```python
# 문제 1-1: 사칙연산

x = int(input("첫 번째 숫자를 입력하세요: "))
y = int(input("두 번째 숫자를 입력하세요: "))

print(f'첫 번째 숫자를 입력하세요: {x}')
print(f'두 번째 숫자를 입력하세요: {y}')

print(f'{x} + {y} = {x+y}')
print(f'{x} - {y} = {x-y}')
print(f'{x} * {y} = {x*y}')
print(f'{x} / {y} = {x/y:.2f}')
```

```python
# 문제 1-2: 원의 넓이와 둘레 계산

import math

radius = int(input("원의 반지름을 입력하세요: "))
print(f'원의 반지름을 입력하세요: {radius}')
print(f'반지름이 {radius}인 원의 넓이: {radius*radius*math.pi:.2f}')
print(f'반지름이 {radius}인 원의 둘레: {radius*2*math.pi:.2f}')
```

```python
# 문제 1-3: 섭씨 온도를 화씨 온도로 변환

temp_c = int(input("섭씨 온도를 입력하세요: "))
temp_f = temp_c * 9/5 + 32

print(f'섭씨 온도를 입력하세요: {temp_c}')
print(f'섭씨 {temp_c}도는 화씨 {temp_f}도 입니다.')
```

```python
# 문제 1-4: 직사각형의 넓이와 둘레 계산

a = int(input("가로 길이를 입력하세요: "))
b = int(input("세로 길이를 입력하세요: "))

print(f'가로 길이를 입력하세요: {a}')
print(f'세로 길이를 입력하세요: {b}')

print(f'직사각형의 넓이: {a*b}')
print(f'직사각형의 둘레: {2*(a+b)}')
```

```python
# 문제 1-5: 상품의 할인된 가격 계산

price = int(input("상품 가격을 입력하세요: "))
discount = int(input("할인율을 입력하세요(%): "))
discount_price = int(price * discount / 100)

print(f'상품 가격을 입력하세요: {price}')
print(f'할인율을 입력하세요(%): {discount}')

print(f'원래 가격: {price}원')
print(f'할인율: {discount}%')
print(f'할인 금액: {discount_price}원')
print(f'최종 가격: {price - discount_price}원')
```
