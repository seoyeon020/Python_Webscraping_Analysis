```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from IPython.display import Image, display

section_dict = {
    '최신': 'n0100',
    '정치': 'n0200',
    '경제': 'n0300',
    '사회': 'n0400',
    '세계': 'n0500',
    'IT/과학': 'n0600'
}

def print_news(section_name):
    mid = section_dict.get(section_name, 'n0100')

    # URL 설정 (최신뉴스와 다른 섹션의 url이 다름)
    if mid == 'n0100':
        url = f'https://news.nate.com/recent?mid={mid}'
    else:
        url = f'https://news.nate.com/section?mid={mid}'
    
    print(f'====> {url} {section_name} 뉴스 <====')

    req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
    }

    res = requests.get(url, headers=req_header)
    res.encoding = 'euc-kr'

    if res.ok:
        soup = BeautifulSoup(res.text, 'html.parser')
        news_list = soup.select('div.mlt01')

        for tags in news_list:
            a_tag = tags.find('a')

            if a_tag:
                link = a_tag['href']
                if link.startswith('//'):
                    link = 'https:' + link

                title_tag = tags.select_one('h2.tit')
                if title_tag:
                    title = title_tag.text.strip()
                    print(link)
                    print(title)

                    # 이미지 출력  
                    img_tag = tags.select_one('img')
                    if img_tag and img_tag.get('src'):
                        src = img_tag['src']
                        # urljoin을 사용하여 도메인과 경로 합치기
                        full_img_url = urljoin(url, src)
                        # display로 이미지 출력 
                        display(Image(url=full_img_url))
                    else:
                        print("이미지 없음")

    else:
        print(f'Error Code = {res.status_code}')
```

```python
print_news('최신')
```

```python
import requests
from bs4 import BeautifulSoup
import os

def download_one_episode(title, no, url):
    
    req_header = {
        'referer': url,
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.3'
    }

    # 1. img\title\no 경로 설정
    imgdir_path = os.path.join('img', title, str(no))
    # 디렉터리 없으면 생성
    os.makedirs(imgdir_path, exist_ok=True)

    # 2. 웹페이지 요청 
    res = requests.get(url, headers=req_header)
    if not res.ok:
        print(f'Error Code = {res.status_code}')
        return
    
    soup = BeautifulSoup(res.text, 'html.parser')
    img_url_list = [img_tag['src'] for img_tag in soup.select("img[src*='IMAG01']")]

    # 이미지 다운로드
    for img_url in img_url_list:
        res = requests.get(img_url, headers=req_header)

        if res.ok:
            img_data = res.content
            file_name = os.path.basename(img_url)
            # 최종 저장 경로 (img\제목\번호\파일명.jpg)
            file_path = os.path.join(imgdir_path, file_name)

            with open(file_path, 'wb') as file:
                file.write(img_data)
                print(f'저장 완료: {file_name}')
        else:
            print(f'다운로드 실패: {img_url}')
```

```python
download_one_episode('육아일기 시즌2', 77, 'https://comic.naver.com/webtoon/detail?titleId=812354&no=212&week=sun')
```

```python
download_one_episode('아홉수 우리들', 280, 'https://comic.naver.com/webtoon/detail?titleId=724815&no=283&week=sat')
```
