#### 4-1 인구현황 데이터 분석

```python
import pandas as pd
person = pd.read_csv('../data/인구현황.csv')
person = person[person['행정기관'] != '전국']
person
```

```python
# 1. 지역별 총인구수 Top5
top5 = person[['행정기관','총인구수']].sort_values(by='총인구수',ascending=False)
top5.head()
```

```python
# 1-1. 지역별 총인구수 plot 시각화

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

#한글폰트 path 설정
font_path = 'C:\\Windows\\Fonts\\malgun.ttf'
#font의 파일정보로 font name 을 알아내기
font_name = fm.FontProperties(fname=font_path).get_name()
matplotlib.rc('font', family=font_name)

figure, ax1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(18,9)
top5_plot = sns.barplot(data=top5, x="행정기관", y="총인구수", ax=ax1,hue='행정기관')
top5_plot.set_title(f'지역별 총인구수')
plt.xticks(rotation=45)

plt.show()
```

```python
# 2. 세대당 인구수 높은 지역 To5 

house5 = person[['행정기관','세대당 인구']].sort_values(by='세대당 인구',ascending=False).reset_index(drop=True)
house5.head()
```

```python
# 2-1. 지역별 세대당 인구 Plot

figure, ax1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(18,9)
house5_plot = sns.barplot(data=house5, x="행정기관", y="세대당 인구", ax=ax1,hue='행정기관')
house5_plot.set_title(f'지역별 세대당 인구')
plt.xticks(rotation=45)

plt.show()
```

```python
# 3. 남녀 비율 분석

fm = person[['행정기관','남여 비율']].sort_values(by='남여 비율',ascending=False).reset_index(drop=True)
fm
```

```python
# 3-1. 남녀 비율 분석 plot

figure, ax1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(18,9)
fm_plot = sns.barplot(data=fm, x="행정기관", y="남여 비율", ax=ax1,hue='행정기관')
fm_plot.set_title(f'지역별 남녀 비율')
plt.xticks(rotation=45)

plt.show()
```
