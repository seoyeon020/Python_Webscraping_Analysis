```python
import requests
from bs4 import BeautifulSoup

# 1. URL 설정
url = 'https://news.daum.net/economy'
print(url)

# 2. 요청 헤더 설정
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
}

# 3. requests로 가져오기
res = requests.get(url, headers=req_header)
res.encoding = 'utf-8'

print(type(res))
print(res.status_code)

if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')
    # 4. ul 태그 가져오기
    news_list = soup.select('ul.list_newsheadline2 li')
    print(type(news_list), len(news_list))
    # 5. 반복문으로 링크와 제목 추출
    for li_tag in news_list:
        a_tag = li_tag.find('a')
        if a_tag:
            link = a_tag['href']

            # 제목 태그(strong)
            title_tag = li_tag.select_one('strong.tit_txt')
            if title_tag:
                title = title_tag.text.strip()
                print(link)
                print(title)
else:
    print(f'Error Code = {res.status_code}')
```

```python
import requests
from bs4 import BeautifulSoup

section_dict = {'기후/환경':'climate','사회':'society','경제':'economy','정치':'politics','국제':'world','문화':'culture','생활':'life','IT/과학':'tech','인물':'people'}

def print_news(section_name):
    sid = section_dict.get(section_name,'사회')
    url = f'https://news.daum.net/{sid}'
    print(f'====> {url} {section_name} 뉴스 <====')

    req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
    }

    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'

    if res.ok:
        soup = BeautifulSoup(res.text, 'html.parser')
        news_list = soup.select('ul.list_newsheadline2 li')
        
        for li_tag in news_list:
            a_tag = li_tag.find('a')
            if a_tag:
                link = a_tag['href']

                title_tag = li_tag.select_one('strong.tit_txt')
                if title_tag:
                    title = title_tag.text.strip()
                    print(link)
                    print(title)
    else:
        print(f'Error Code = {res.status_code}')
```

```python
print_news('생활')
```
